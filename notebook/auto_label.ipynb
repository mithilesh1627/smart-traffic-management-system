```python
from ultralytics import YOLO
from pathlib import Path

DATASET_DIR = Path('D:\2026\CV_Project\smart-traffic-management-system\IDD_Dataset')
TRAIN_IMG_DIR =DATASET_DIR/"Processed_dataset"/"train"/"images"
CONF_THRES =0.6
```

```python
model = YOLO(model='yolo11n.pt',task='detect',verbose=True)
```

```python
label = DATASET_DIR/"Processed_dataset"/"train"/"labels"
```

```python

for img_path in TRAIN_IMG_DIR.rglob('*.jpg'):
    results = model(img_path,conf=CONF_THRES)


    for r in results:
        if r.boxes is None or len(r.boxes)==0:
            continue

        label_path = label / img_path.relative_to(TRAIN_IMG_DIR)
        label_path = label_path.with_suffix(".txt")
        label_path.parent.mkdir(parents=True, exist_ok=True)

        lines = []

        for box in r.boxes:
            cls = int(box.cls)
            x,y,w,h = box.xywhn[0].tolist()
            lines.append(
                f"{cls} {x:.6f} {y:.6f} {w:.6f} {h:.6f}"
            )
        
        with open(label_path,'w') as f:
            f.write('\n'.join(lines))
            
```
